```python
import torchvision
from torchvision import transforms

from torch.utils.data import DataLoader
```

```python
from matplotlib.pyplot import imshow
%matplotlib inline
```

```python
trans = transforms.Compose([
    transforms.Resize((64, 128))
])

train_data = torchvision.datasets.ImageFolder(root='C:/Users/jys74/Desktop/modeep/origin_data', transform=trans)
```

```python
for num, value in enumerate(train_data):
    data, label = value
    print(num, data, label)

    if(label == 0):
        data.save('C:/Users/jys74/Desktop/modeep/origin_data/gray/%d_%d.jpeg'%(num,label))
    else:
        data.save('C:/Users/jys74/Desktop/modeep/origin_data/red/%d_%d.jpeg'%(num,label))
```
